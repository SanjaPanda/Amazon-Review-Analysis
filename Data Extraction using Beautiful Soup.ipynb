```python
#import important libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time as time
```

```python
#Product Data

product_details = pd.DataFrame(columns = ['asin','product_name','price','review_url'])
product_details.head()
```

```python
#Review Data

review_details = pd.DataFrame(columns = ['asin','title','review','star'])
review_details.head()
```

```python
#Setting base variables

search_query="bluetooth+pods"
base_url="https://www.amazon.in/s?k="

headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36", 
           'referrer': 'https://google.com',
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
            'Accept-Encoding': 'gzip, deflate, br',
            'Accept-Language': 'en-US,en;q=0.9',
            'Pragma': 'no-cache'}  #To get header; google search 'what is my user agent'

delays = [7, 4, 6, 2, 10, 19]
```

```python
#Get ASIN number to identify products

pageno = '&page='
data_asin = []

for i in range(1,5):
    url=base_url+search_query+pageno+str(i)
    search_response=requests.get(url,headers = headers)
    delay = np.random.choice(delays)
    time.sleep(delay)
    soup=BeautifulSoup(search_response.content,'html.parser')
    for j in soup.findAll('div',{'data-component-type':'s-search-result'}):
        data_asin.append(j['data-asin'])
                
data_asin = list(set(data_asin))
```

```python
data_asin
```

```python
#Get product data, also review link to fetch product reviews

product_url = 'https://www.amazon.in/dp/'
review = []

for i in data_asin:
    url = product_url+i
    product_response=requests.get(url,headers = headers)
    delay = np.random.choice(delays)
    time.sleep(delay)
    if product_response.status_code == 200:
        soup=BeautifulSoup(product_response.content,'html.parser')
        
        for j in soup.findAll('body'):
            product = ''
            price = ''
            review_link = ''
            print(url)

            try:
                product = j.find('span',{'id':'productTitle'}).get_text().strip()
                try:
                    price = j.find('span',{'id':'priceblock_dealprice'}).get_text().strip()
                except:
                    pass
                try:
                    review_link = "https://www.amazon.in/"+j.find('a',{'data-hook':'see-all-reviews-link-foot'})['href']
                    review.append(review_link)
                except:
                    pass

                product_details = product_details.append({'asin': i,
                                                      'product_name':product,
                                                      'price':price,
                                                      'review_url':review_link},ignore_index=True)
            except:
                pass
```

```python
#Scrapping Reviews

pgno = '%2F&pageNumber='
for i,rev in enumerate(review):
    for k in range(1,10):
        url = rev+pgno+str(k)
        review_response=requests.get(url,headers = headers)
        time.sleep(delay)
        soup=BeautifulSoup(review_response.content,'html5lib')
        if review_response.status_code == 200:
            print(url)
            star = 0
            title = ''
            review_data = ''
            for j in soup.findAll('div',{'data-hook':'review'}):
                #print(j)
                try:
                    star = j.find('i',{'data-hook':'review-star-rating'}).get_text().strip()[0:3]
                    title = j.find('a',{'data-hook':'review-title'}).get_text()
                    review_data = j.find('span',{'data-hook':'review-body'}).get_text()

                    review_details = review_details.append({'asin':data_asin[i],
                                   'title':title,
                                   'review':review_data,
                                   'star':star},ignore_index=True)
                except:
                    pass
```
